```python
# Filter rows with missing values in 'payment_value'
missing_payment_value = df[df['payment_value'].isnull()]
missing_payment_value
```

```python
def fill_missing_payment_value(df):
  """
  Fills missing values in payment_value based on price, order_item_id, and freight_value

  Args:q
    df: The DataFrame containing the data

  Returns:
    DataFrame: The DataFrame with filled missing values
  """

  df['payment_value'] = df['payment_value'].fillna(df.apply(lambda row: row['price'] * row['order_item_id'] + row['freight_value'], axis=1))
  return df

df = fill_missing_payment_value(df)
```

### B3. payment_type and payment_installments

```python
def count_payment_types(df):
  """Counts unique payment_type and their totals

  Args:
    df: The DataFrame containing the payment_type column

  Returns:
    pandas.Series: A Series with payment types as index and counts as values
  """

  payment_type_counts = df['payment_type'].value_counts()
  return payment_type_counts

payment_type_counts = count_payment_types(df)
print(payment_type_counts)
```

```python
# Counts unique payment_installments and their totals
def count_payment_installments(df):

  payment_installments_counts = df['payment_installments'].value_counts()
  return payment_installments_counts

payment_installments_counts = count_payment_installments(df)
print(payment_installments_counts)
```

```python
def fill_missing_with_mode(df):
  """
  Fills missing values in payment_type and payment_installments with their mode

  Args:
    df: The DataFrame containing the data

  Returns:
    DataFrame: The DataFrame with filled missing values
  """

  for col in ['payment_type', 'payment_installments']:
    mode_value = df[col].mode().iloc[0]  # Get the mode
    df[col].fillna(mode_value, inplace=True)

  return df
df = fill_missing_with_mode(df)
```

## A. order_status condition

```python
def analyze_order_status(df):
  """
  Counts unique order statuses, calculates percentages, and plots the results.

  Args:
    df: The DataFrame containing the order_status column.
  """

  # Count unique order statuses
  order_status_counts = df['order_status'].value_counts()

  # Calculate percentages
  order_status_percentages = (order_status_counts / order_status_counts.sum()) * 100

  # Create a DataFrame with counts and percentages
  order_status_summary = pd.DataFrame({'Count': order_status_counts, 'Percentage': order_status_percentages})

  # Plot the counts
  order_status_counts.plot(kind='bar', title='Order Status Counts')
  plt.ylabel('Count')
  plt.xlabel('Order Status')

  plt.show()

  return order_status_summary

# Example usage:
# Assuming you have a DataFrame named 'df'
analyze_order_status(df)
```

```python
# Canceled and unavailable in order_status will be drop due to not affected in the performance transaction or sales in generate revenue

def filter_orders(df):
  """
  Filters a DataFrame to remove rows with 'canceled' or 'unavailable' order status.

  Args:
    df: The DataFrame containing the order data.

  Returns:
    DataFrame: The filtered DataFrame.
  """

  return df[~df['order_status'].isin(['canceled', 'unavailable'])]

# Example usage:
# Assuming you have a DataFrame named 'df'
df = filter_orders(df)
df.shape
```

Orders with canceled or unavailable statuses will be dropped due to their negligible impact on transaction performance and revenue generation.

# 4. RFM or Clustering

## A. Recency

```python
# Check the date
print("Min Date", df["order_purchase_timestamp"].min(), "Max Date", df["order_purchase_timestamp"].max())
```

```python
# Convert column order_purchase_timestamp to datetime format
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])

# Calculate total last days customer made a purchase or transaction
recency = (dt.datetime(2018, 9, 3) - df.groupby("customer_unique_id").agg({"order_purchase_timestamp":"max"})).rename(columns = {"order_purchase_timestamp":"Recency"})
recency["Recency"] = recency["Recency"].apply(lambda x: x.days)

# Reset the index to make 'customer_unique_id' a column
recency = recency.reset_index()
recency
```

## B. Frequency

```python
# Calculate the frequency customer made a purchase or transaction by their order_id
freq = df.groupby("customer_unique_id").agg({"order_id":"nunique"}).rename(columns={"order_id": "Frequency"})
freq = freq.reset_index()
freq
```

## C. Monetary

```python
# Calculate the total payment_value by each customer_unique_id
monetary = df.groupby("customer_unique_id").agg({"payment_value":"sum"}).rename(columns={"payment_value":"Monetary"})
monetary = monetary.reset_index()
monetary
```

```python
# Merge recency, freq, and monetary DataFrames
rfm = pd.merge(recency, freq, on='customer_unique_id')
rfm = pd.merge(rfm, monetary, on='customer_unique_id')

# Now merged_df contains columns from all three DataFrames
rfm
```

```python
# Create RFM Score
rfm["Recency_Score"] = pd.qcut(rfm["Recency"], 5, labels = [5, 4 , 3, 2, 1])
rfm["Frequency_Score"]= pd.qcut(rfm["Frequency"].rank(method="first"),5, labels=[1,2,3,4,5])
rfm["Monetary_Score"] = pd.qcut(rfm['Monetary'], 5, labels = [1, 2, 3, 4, 5])
rfm
```

```python
# Define RFM Scores: Category
rfm["RFM_SCORE"] = (rfm['Recency_Score'].astype(str) + 
                     rfm['Frequency_Score'].astype(str) + 
                     rfm['Monetary_Score'].astype(str))
rfm
```

```python
# Cluster the segmentation of each customer_unique_id
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Can\'t Loose',
    r'3[1-2]': 'About to Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}

rfm['Segment'] = rfm['Recency_Score'].astype(str) + rfm['Frequency_Score'].astype(str)
rfm['Segment'] = rfm['Segment'].replace(seg_map, regex=True)
rfm
```

Based on Recency (R)
- Hibernating: Customers who haven't made a purchase in a long time but have previously been active.
- At Risk: Customers who haven't made a purchase recently, but their previous activity suggests they might churn if not re-engaged.
- Can't Lose: Customers who have made a purchase recently but have a low overall purchase frequency.

Based on Frequency (F)
- About to Sleep: Customers who have made a purchase recently but have a declining frequency compared to their past.
- Need Attention: Customers who have made a purchase recently but their frequency is significantly lower than their historical average.

Based on Monetary (M)
- Loyal Customers: Customers who have made multiple purchases and have spent a significant amount of money.
- Promising: Customers who have made a few purchases but have spent a significant amount of money.
- New Customers: Customers who have made their first purchase.
- Potential Loyalists: Customers who have made a few purchases and have shown consistent spending behavior.
- Champions: Customers who have made multiple purchases, spent a significant amount of money, and are likely to continue purchasing.
